```python
# import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load DataSet
flower = sns.load_dataset("iris")
flower.head()

#draw a Line Pilot

sns.lineplot(x="sepal_length" , y="sepal_width", data=flower)
plt.title("Iris Plot")
plt.show()
```

# Adding Limits

```python
# import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load DataSet
flower = sns.load_dataset("iris")
flower.head()

#draw a Line Pilot

sns.lineplot(x="sepal_length" , y="sepal_width", data=flower)
plt.title("Iris Plot")
plt.xlim(5)
plt.ylim(3)
plt.show()
```
