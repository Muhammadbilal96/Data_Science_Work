# One-sample student's t-test

Test a sample with a known standard value. 

- **Assumptions**

- Observations in each sample are independent and identically distributed.

- Observations in each sample are normally distributed.

- **Interpretation**

>*H0:* the means of the samples are equal to the known value.

>*H1:* the means of the samples are unequal to the known value.


```python
# One Sample T-Test

#import Libraries
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_1samp

# Import DataSet

df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df1 = df[['sex','age','fare']]
df1.head()
```

```python
# Data Describe
df1.describe()
```

```python
# Check The age and comapre with a know value of 45 years

ttest_1samp(df1['age'],50)
stat , p = ttest_1samp(df1['age'],50)

print('stat=%.3f , p=%.3f' % (stat , p))

# Adding Conditional Argument For Ease
if p > 0.05:
    print("Probably The Same Distribution")
else:
    print("Probably Different Distribution")    
```

# Two Sample T-Test
## Independent student's t-test

- **Assumptions**
- Observations in each sample are independent and identically distributed.

- Observations in each sample are normally distributed.

- Observations in each sample have the same variance.

- **Interpretation**

>*H0:* the means of the samples are equal.

>*H1:* the means of the samples are unequal.

```python
# We Will Compare age and fare of male vs Female Passenger

#Splitting Dataset

df_male = df1.loc[df1['sex']=='male']
df_female = df1.loc[df1['sex']=='female']

# Import Libraries
from scipy.stats import ttest_ind

stat , p = ttest_ind(df_male['fare'], df_female['fare'])
print('stat=%.3f , p=%.3f' % (stat , p))

# Adding Conditional Argument For Ease
if p > 0.05:
    print("Probably The Same Distribution")
else:
    print("Probably Different Distribution")  
```


# Paired student's t-test

Tests whether the means of two paired samples are significantly different.

- **Assumptions**

- Observations in each sample are independent and identically distributed.

- Observations in each sample are normally distributed.

- Observations in each sample have the same variance.

- Observations across each sample are paired.

- **Interpretation**

>*H0:* the means of the samples are equal.

>*H1:* the means of the samples are unequal.
