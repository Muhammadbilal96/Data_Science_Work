# Line Plot


# Import Lib


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# import data set

```python
df = sns.load_dataset("iris")
df

sns.lineplot(x="sepal_length" , y="sepal_width", data=df)
plt.show()
```

# Adding Title

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")

sns.lineplot(x="sepal_length" , y="sepal_width" , data=df)
plt.title("Phool Ka Graph")
plt.show()
```

# Adding Limits

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")

sns.lineplot(x="sepal_length" , y="sepal_width" , data=df)
plt.title("Phool Ka Graph")
plt.xlim(5.5)
plt.ylim(2.8)
plt.show()
```

# Set Style

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")

sns.lineplot(x="sepal_length" , y="sepal_width" , data=df)
plt.title("Phool Ka Graph")
sns.set_style("darkgrid")
sns.set_style(style=None, rc=None)
plt.show()
```

```python


# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None, rc=None)

# Load dataset
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Dark set of iris")
sns.set_style('dark')
```

# Size of Figure

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt


# Load dataset
phool = sns.load_dataset("iris")

# Size Of Figure
plt.figure(figsize=(8,4))

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Dark set of iris")
```
