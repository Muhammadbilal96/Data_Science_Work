```python
# import libaray 
import pandas as pd
import seaborn as sns
import numpy as np

# load Data Set
df = sns.load_dataset("titanic")

```

```python
# Plot use
sns.boxplot(x = "survived", 
            y="age",showmeans=True,
            meanprops={"marker" : "*",
                       "markersize" : "14",
                       "markeredgecolor" : "blue"}, 
            data=df)
```

```python
# Plot use
sns.boxplot(x = "survived", 
            y="age",showmeans=True,
            meanprops={"marker" : "*",
                       "markersize" : "14",
                       "markeredgecolor" : "blue"}, 
            data=df)
plt.xlabel("How Many Survied"),
plt.ylabel("Age(Year)"),
plt.title("Plot Ko Dikhlo Bhai Jan")

plt.show()
```
