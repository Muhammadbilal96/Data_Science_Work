##                                  (Logistic Regression & Image Classification)

```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Data Set

from sklearn.datasets import load_digits

digits = load_digits()
```

```python
digits.data.shape

X = digits.data
```

```python
digits.target.shape

y = digits.target
```

```python
plt.figure(figsize=(20,4))

for index , (image , label) in enumerate(zip(digits.data[0:10] , digits.target[0:10])):
    plt.subplot(1,10 , index + 1)
    plt.imshow(np.reshape(image , (8,8)), cmap=plt.cm.gray)
    plt.title('Training : %i\n' % label , fontsize=20)
```

```python
plt.figure(figsize=(20,4))

for index , (image , label) in enumerate(zip(digits.data[0:20] , digits.target[0:20])):
    plt.subplot(1,20 , index + 1)
    plt.imshow(np.reshape(image , (8,8)), cmap=plt.cm.gray)
    plt.title(label , fontsize=20)
```

```python
plt.figure(figsize=(20,4))

for index , (image , label) in enumerate(zip(digits.data[0:30] , digits.target[0:30])):
    plt.subplot(1,30 , index + 1)
    plt.imshow(np.reshape(image , (8,8)), cmap=plt.cm.gray)
    plt.title(label , fontsize=20)
```

```python
# Splite Data & Train Data

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , train_size=0.2 , random_state=0)
```

```python
print("Train input Data" , X_train.shape)
print("Train input Data" , X_test.shape)
print("Train output Data" , y_train.shape)
print("Train output Data" , y_test.shape)
```

```python
# Train Model

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression().fit(X , y)

log_reg
```

```python
log_reg.predict(X_test[0:10])
```

```python
# Accuancy Check

score = log_reg.score(X_test , y_test)

print("The Accuaracy Score Is"  , score)
```
