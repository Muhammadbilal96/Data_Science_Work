### 9 Step Check The Normality 

```python
sns.histplot(df['age'])
```

```python
# Shapiro Wilk Test

from scipy.stats import shapiro

stats , p_value = shapiro(df['age'])

print("stats=%.3f , p_value=%.3f" % (stats,p_value))

if p_value > 0.05:

    print("Data Is Normality / Gussain Distribution")

else:

     print("Data Is Not Normality / Gussain Distribution")    
```

```python
sns.histplot(df, x="age", hue="who", multiple="dodge")
```

```python
# Measure its skewness and kurtosis

df['age'].agg(['skew','kurtosis']).transpose()
```
