## 16 Splitting a string into multipe columns 

```python
import pandas as pd
```

```python
df = pd.DataFrame({'name':['Muhammad Bilal','Ahmed Raza','Aqsa Ahmed','Haider Ali'], 
                   'Location':['Riyaz , Saudia Arab','Lahore , Pakistan','Abu Dabi , UAE','Karachi , Pakistan']})
df
```

```python
# spliting a column into two column

df[["first_name","last_name"]] = df.name.str.split(' ',expand=True)
df
```

```python
df[["city","country"]] = df.Location.str.split(', ',expand=True)
df
```

```python
# Refine data Manupulation

df = df[['first_name','last_name','city','country']]
df
```
