```python
# Find Outliers

Q1 = clean_df['age'].quantile(0.25)
Q3 = clean_df['age'].quantile(0.75)

Q1 , Q3
```

```python
IQR = Q3 - Q1

IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q1 + 1.5*IQR

lower_limit , upper_limit
```

```python
clean_df[(clean_df['age']>upper_limit)]
```

```python
clean_df[(clean_df['age']<lower_limit)]
```

```python
data = clean_df
```

```python
data['age'] = np.where(
    data['age'] > upper_limit,
    upper_limit,
    np.where(
    data['age'] < lower_limit,
    lower_limit,
    data['age']
    )
)
```

```python
sns.boxplot(y='age' , data=data)
```

```python
sns.distplot(data['age'])
```
